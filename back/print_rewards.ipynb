```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
```

```python
df1 = pd.read_csv('PPO_logs1/Splendor/PPO_Splendor_log_13.csv', index_col=0)
df2 = pd.read_csv('PPO_logs2/Splendor/PPO_Splendor_log_13.csv', index_col=0)
```

```python
y_smooth1 = savgol_filter(df1["reward"], 10, 2)
y_smooth2 = savgol_filter(df2["reward"], 10, 2)
#y_smooth1 = df1["reward"]
#y_smooth2 = df2["reward"]
```

```python
# Crer un nouveau graphique
fig, ax = plt.subplots()

# Ajouter la courbe bleue
ax.plot(df1["timestep"], y_smooth1, color="blue", label=" Player 1")

# Ajouter la courbe verte
ax.plot(df2["timestep"], y_smooth2, color="green", label="Player 2")

# Ajouter des lgendes
ax.set_xlabel("Timestep")
ax.set_ylabel("Reward")
ax.set_title("Rcompense en fonction du temps")
ax.legend()

# Afficher le graphique
plt.show()
```
